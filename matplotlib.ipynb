```python
# Plot Types
# 1. Line Plot/Chart- used to represent the relation between two data X and Y on a different axis
```

```python
fig=plt.figure() # Create a new figure, or activate an existing figure.
ax=plt.axes() #Add an Axes to the current figure and make it the current Axes.
X=np.linspace(0,10,100) #Return evenly spaced numbers over a specified interval.
plt.plot(X,np.sin(X),'r-')
```

```python
#Scatter Plot- The scatter() function plots one dot for each observation. 
#It needs two arrays of the same length, one for the values of the x-axis, and one for values on the y-axis

fig=plt.figure() # Create a new figure, or activate an existing figure.
ax=plt.axes() #Add an Axes to the current figure and make it the current Axes.
X=np.linspace(0,10,100) #Return evenly spaced numbers over a specified interval.
plt.scatter(X,np.sin(X))
```

```python
# Scatter plot using only Plot function
fig=plt.figure() # Create a new figure, or activate an existing figure.
ax=plt.axes() #Add an Axes to the current figure and make it the current Axes.
X=np.linspace(0,10,100) #Return evenly spaced numbers over a specified interval.
plt.plot(X,np.sin(X),'ro')
```

```python
# Histogram-Graph showing frequency distribution
#graph showing the number of observations within each given interval.

X=np.random.randn(1000) #Return a sample (or samples) from the "standard normal" distribution.
plt.grid()
plt.hist(X,color='green') #Compute and plot a histogram.


```

```python
#Bar Chart/Plot-Graph that represent the category of data with rectangualr bar with length and height 
#that is propostional to values which they represent
# It can be plotted horizontally/Vertically
#Describe the comparisons between descrete categories

X=[5,25,50,20]
plt.bar(range(len(X)),X,color='red')
```

```python
#Horizontal Bar Plot
X=[5,25,50,20]
plt.barh(range(len(X)),X,color='red')
```

```python
#Error Bar Chart-Graphical enhancemnet that visualizes the variability of plotted data on a cartisian graph
# Can be applied to graph to provide an additional layer of detail on the presented data
# indicate estimatted error or uncertainity 
#A short error bar shows that values are concentrated signaling that the plotted averaged value is more likely 
#while a long error bar would indicate that the values are more spread out and less reliable.

X=np.arange(0,4,0.2)
Y=np.exp(-X) #Calculate the exponential of all elements in the input array.
Z=0.1*np.abs(np.random.randn(len(Y)))
# Abs- Calculate the absolute value element-wise.
plt.errorbar(X,Y,yerr=Z,fmt='.-',color='red')
plt.show()
```

```python
# Stacked Bar Chart/Plot-extends the standard bar chart from looking at numeric values accross one categorical variable to two
# Each bar in a standard bar chart is divided into a number of sub-bars stacked end to end, 
#each one corresponding to a level of the second categorical variable.
#objective of a standard bar chart is to compare numeric values between levels of a categorical variable
X=[45,20,50,30]
Y=[10,30,5,60]
Z=range(4)
plt.bar(Z,X,color='red')
plt.bar(Z,Y,color='black',bottom=X)
```

```python
#Pie Chart-Circular statisical plot that can display only one series of data
# Area of chart is total percentage of given data

X=[10,30,50,5]
Y=['grocery','Electronic','Apprael','Pharma']
plt.pie(X,labels=Y,shadow=True)
plt.show()
```

```python
#Box Plot-created to display the summary of the set of data values having properties like minimum, first quartile, median, third quartile and maximum
# Also called whisker plot
# x-axis denotes the data to be plotted while the y-axis shows the frequency distribution
X=np.random.randn(100)
plt.boxplot(X)
            
plt.show()
```

```python
#Area Chart-really similar to a line chart, except that the area between the x axis and the line is filled in with color or shading
#It represents the evolution of a numeric variable.
# Two main ways to build area chart- Fillbetween() and Stackplot()
X=range(1,6)
Y=[1,4,6,8,4]
# Area Plot
plt.fill_between(X,Y,color='red')
plt.show()
```

```python

# Area plot using stackplot
X=range(1,6)
Y=[1,4,6,8,4]
# Area Plot
plt.stackplot(X,Y)
plt.show()
```

```python
#Contour Plot-a tool for doing multivariate analysis and visualizing 3-D plots in 2-D space
# Also called levels plot
con1=np.random.rand(10,20)
plt.contour(con1)
plt.show()
```

```python
#Addition Grid
X=np.arange(1,6)
plt.plot(X,X*2,X,X*3,X*1.4,X*.5)
plt.grid(True) #Configure the grid lines
plt.show()
```

```python
#Axis Handling
X=np.arange(1,6)
plt.plot(X,X*2,X,X*3,X,X*1.4,X,X*.5)
plt.grid(True) #Configure the grid lines
plt.axis([0,8,0,19])
plt.show()
```

```python
X=np.arange(1,6)
plt.plot(X,X*2,X,X*3,X,X*1.4,X,X*.5)
plt.xlim(1,6)#Get or set the x limits of the current axes.
plt.ylim(2,10)
plt.show()
```

```python
X=[5,4,6,10,2,9,1,0]
plt.plot(X)
plt.grid()
plt.xticks([2,4,6,10])#Get or set the current tick locations and labels of the x-axis.
plt.yticks([2,4,6,10])
plt.show()
```

```python
# Adding labels and Title
X=[5,4,6,10,2,9,1,0]
plt.plot(X)
plt.grid()
plt.xticks([2,4,6,10])#Get or set the current tick locations and labels of the x-axis.
plt.yticks([2,4,6,10])
plt.xlabel("X Axis")
plt.ylabel("Y Axis")
plt.title("This is Title")
plt.legend("Test")
plt.style.use('_classic_test_patch')
plt.show()
```
